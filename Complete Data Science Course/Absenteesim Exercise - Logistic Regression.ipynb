## standarize the data 

```python
from sklearn.preprocessing import StandardScaler

absenteesim_scaler = StandardScaler()
```

```python
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler

class CustomScaler(BaseEstimator, TransformerMixin):
    
    def __init__(self, columns, copy=True, with_mean=True, with_std=True):
        # Initializing StandardScaler with the passed arguments
        self.scaler = StandardScaler(copy=copy, with_mean=with_mean, with_std=with_std)
        self.columns = columns 
        self.mean_ = None 
        self.var_ = None
        
    def fit(self, X, y=None):
        # Fitting the scaler to the selected columns
        self.scaler.fit(X[self.columns], y)
        # Storing the mean and variance
        self.mean_ = np.mean(X[self.columns], axis=0)
        self.var_ = np.var(X[self.columns], axis=0)
        return self 
    
    def transform(self, X, y=None):
        # Transforming the selected columns using the fitted scaler
        init_col_order = X.columns
        X_scaled = pd.DataFrame(self.scaler.transform(X[self.columns]), columns=self.columns)
        # Keeping the columns that were not scaled
        X_not_scaled = X.loc[:, ~X.columns.isin(self.columns)]
        # Returning the DataFrame with original column order
        return pd.concat([X_not_scaled, X_scaled], axis=1)[init_col_order]
```
